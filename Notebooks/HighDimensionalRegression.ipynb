# Zadanie 1. Dobór regularyzacji dla modeli Lasso
Używając danych z projektu zaliczeniowego nr 2 dla grupy 5, pokaż że domyślne parametry regularyzacji w modelu StdScaler + Lasso, tj. alpha=1.0, dają trywialny klasyfikator. Znajdź metodą prób (bisekcja?) możliwie najlepszą wartość dla <alpha> względem błędu MSE

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.dummy import DummyRegressor
```

```python
X_train = pd.read_csv("5/X_train.csv").astype('float32')
y_train = pd.read_csv("5/y_train.csv").astype('float32').set_index("Id")
```

```python
model = Pipeline(
   [
       ('scaler', StandardScaler()),
       ('predictor', DummyRegressor())
   ]
)


model.fit(X_train,y_train)
mean_squared_error(y_train,model.predict(X_train))
```

```python
# Define the pipeline with Lasso regression
model = Pipeline([
    ('scaler', StandardScaler()),
    ('predictor', linear_model.Lasso(alpha=1.0))  # Initial alpha value
])

# Fit the model
model.fit(X_train, y_train)

# Calculate initial MSE
initial_mse = mean_squared_error(y_train, model.predict(X_train))
print(initial_mse)
```

```python
# Przedział dla alpha
alpha_low = 0.001
alpha_high = 0.01

# Dokładność
epsilon = 0.0001

# Funkcja do obliczania MSE dla danej wartości alpha
def calculate_mse_with_alpha(alpha):
    model = Pipeline([
        ('scaler', StandardScaler()),
        ('predictor', linear_model.Lasso(alpha=alpha))
    ])
    model.fit(X_train, y_train)
    y_pred = model.predict(X_train)
    mse = mean_squared_error(y_train, y_pred)
    return mse

# Algorytm bisekcji do znalezienia optymalnej wartości alpha
def find_optimal_alpha_bisection(alpha_low, alpha_high, epsilon):
    while abs(alpha_high - alpha_low) > epsilon:
        alpha_mid = (alpha_low + alpha_high) / 2
        mse_mid = calculate_mse_with_alpha(alpha_mid)
        mse_low = calculate_mse_with_alpha(alpha_low)
        if mse_mid < mse_low:
            alpha_low = alpha_mid
        else:
            alpha_high = alpha_mid
        print(alpha_low, alpha_high, mse_mid)

    return (alpha_low + alpha_high) / 2

# Znalezienie optymalnej wartości alpha
optimal_alpha = find_optimal_alpha_bisection(alpha_low, alpha_high, epsilon)
print("Optymalna wartość alpha:", optimal_alpha)
print("MSE dla optymalnej wartości alpha:", calculate_mse_with_alpha(optimal_alpha))
```

```python
import numpy as np

def find_optimal_alpha(X, y, lower=1e-5, upper=1e-1, tol=1e-2):
    best_alpha = lower
    best_mse = float('inf')

    while upper - lower > tol:
        alpha = (upper + lower) / 2.0
        model = Pipeline(
            [
                ('scaler', StandardScaler()),
                ('predictor', linear_model.Lasso(alpha=alpha))
            ]
        )
        model.fit(X, y)
        mse = mean_squared_error(y, model.predict(X))

        if mse < best_mse:
            best_mse = mse
            best_alpha = alpha

        if model.predict(X).std() < 1e-5:
            lower = alpha
        else:
            upper = alpha

    return best_alpha, best_mse

optimal_alpha, optimal_mse = find_optimal_alpha(X_train, y_train)
print(f'Optymalna wartość alpha: {optimal_alpha}')
print(f'MSE dla optymalnej wartości alpha: {optimal_mse}')
```
